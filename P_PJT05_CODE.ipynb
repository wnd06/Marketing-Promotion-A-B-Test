### A/B 테스트 진행

---

#### 1. 목표 정의 및 가설 설정
- 목표 : 새로운 마케팅 전략이 기존의 전략보다 더 높은 성과를 이끌어 낼 수 있는지 확인.
- 가설:
    - 귀무가설(H0) : 두 마케팅 전략 간의 성과 차이는 없다. 즉, 새 전략과 기존 전략의 성과가 동일하다.
    - 대립가설(H1) : 새로운 마케팅 전략이 기존 전략보다 더 높은 성과를 보인다.

#### 2. 테스트 기간 설정
- week컬럼에서 1~4로 되어있음. 즉, 한달을 기준으로 확인 할 것.

#### 3. 통계적 분석 수행
- 평균 차이 분석: SalesInThousands의 평균을 비교하여 각 그룹의 성과 차이를 확인합니다.
- T-검정: 두 그룹 간의 성과 차이가 통계적으로 유의미한지 확인하기 위해 독립 표본 T-검정을 수행합니다.

#### Promotion 1과 Promotion 2 성과 차이 분석

````python
import pandas as pd
import scipy.stats as stats

# A/B 그룹 나누기
group_a = df[df['Promotion'] == 1]
group_b = df[df['Promotion'] == 2]

# 평균 차이 검정
t_stat, p_value = stats.ttest_ind(group_a['SalesInThousands'], group_b['SalesInThousands'])

print(f'group_a 평균: {group_a['SalesInThousands'].mean()}')
print(f'group_b 평균: {group_b['SalesInThousands'].mean()}')
print(f"T-statistic: {t_stat}, P-value: {p_value}")
````

#### 분석
- T-statistic이 6.45로 매우 높은 값을 가지고 있습니다. 이는 두 그룹 간의 평균 차이가 상당히 크다는 것을 의미합니다.
- P-value가 3.55e-10으로 매우 낮은 값으로 통계적으로 유의미하다는 것을 의미
- 즉, p-value < 0.05이므로 귀무가설을 기각하고 대립가설을 채택할 수 있습니다.

#### 결론
- 기존 전략이 더 높은 성과를 보였다: Group A(기존 전략)의 평균 매출이 Group B(새 전략)보다 훨씬 높은 것으로 나타났습니다. 또한, 통계적으로 유의미한 차이가 있음을 확인했습니다.

- 추천: 새 전략(B 그룹)은 기존 전략(A 그룹)보다 낮은 성과를 보였기 때문에, 기업 XYZ는 기존 전략을 유지하거나 새 전략을 개선해야 할 필요가 있을 것으로 보입니다.

#### Promotion 1과 Promotion 3 성과 차이 분석

````python
# A/B 그룹 나누기
group_a = df[df['Promotion'] == 1]
group_b = df[df['Promotion'] == 3]

# 평균 차이 검정
t_stat, p_value = stats.ttest_ind(group_a['SalesInThousands'], group_b['SalesInThousands'])

print(f'group_a 평균: {group_a['SalesInThousands'].mean()}')
print(f'group_b 평균: {group_b['SalesInThousands'].mean()}')
print(f"T-statistic: {t_stat}, P-value: {p_value}")
````

#### 분석
- T-statistic 값은 1.55로, 두 그룹 간의 평균 차이가 어느 정도 있지만 큰 차이는 없다는 것을 의미합니다.
- P-value는 0.12로 0.05보다 큰 값으로 통계적으로 유의미하지 않습니다.
- 즉, 귀무가설을 기각 할 수 없고 새 전략과 기존 전략 간의 차이가 우연에 의한 것일 가능성이 큽니다.

#### 결론
- 두 전략 간의 성과 차이는 통계적으로 유의미하지 않다고 볼 수 있습니다.
- Group A (기존 전략)와 Group B (새 전략)의 평균 차이가 있긴 하지만, 그 차이가 우연에 의한 것일 수 있기 때문에, 새 전략이 기존 전략보다 더 효과적이라고 결론 내리기 어렵습니다.

#### Promotion 2와 Promotion 3 성과 차이 분석

````python
# A/B 그룹 나누기
group_a = df[df['Promotion'] == 2]
group_b = df[df['Promotion'] == 3]

# 평균 차이 검정
t_stat, p_value = stats.ttest_ind(group_a['SalesInThousands'], group_b['SalesInThousands'])

print(f'group_a 평균: {group_a['SalesInThousands'].mean()}')
print(f'group_b 평균: {group_b['SalesInThousands'].mean()}')
print(f"T-statistic: {t_stat}, P-value: {p_value}")
````

#### 분석
- T-statistic 값은 -4.88로 매우 낮습니다. 이는 두 그룹 간 차이가 큰 차이가 있는 것을 의미합니다.
- P-value는 1.56e-06로 매우 작습니다. 그러므로 두 그룹 간의 차이는 통계적으로 유의미 한 것으로 보입니다.
- 즉, 귀무가설을 기각하고 대립가설을 채택할 수 있습니다. 그러므로 새 전략이 기존 전략보다 더 좋은 성과를 보인는 것입니다.

#### 결론
- 새 전략이 기존 전략보다 더 나은 성과를 보였음을 알 수 있습니다.
- T-statistic이 매우 크고 P-value가 매우 작기 때문에, 새 전략(B 그룹)이 통계적으로 유의미하게 더 나은 성과를 보였다고 확신할 수 있습니다.
