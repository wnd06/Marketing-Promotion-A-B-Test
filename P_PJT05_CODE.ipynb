````python
def age_change(x):
    if x < 6:
        return '0 ~ 5'
    elif x < 11:
        return '6 ~ 10'
    elif x < 16:
        return '11 ~ 15'
    elif x < 21:
        return '16 ~ 20'
    elif x < 26:
        return '21 ~ 25'
    else :
        return '26 ~ 30'
````

````python
df['Age'] = df['AgeOfStore'].apply(age_change)
````

````python
df[df['AgeOfStore'] == 28]
````

````python
age_sales = df.groupby('Age')['SalesInThousands'].mean().reset_index()
````

````python
age_sales['Age2'] = age_sales['Age'].apply(lambda x : x.split(' ')[0]).astype(int)
age_sales.sort_values('Age2', inplace=True)
````

### A/B ÌÖåÏä§Ìä∏ ÏßÑÌñâ

---

#### 1. Î™©Ìëú Ï†ïÏùò Î∞è Í∞ÄÏÑ§ ÏÑ§Ï†ï
- Î™©Ìëú : ÏÉàÎ°úÏö¥ ÎßàÏºÄÌåÖ Ï†ÑÎûµÏù¥ Í∏∞Ï°¥Ïùò Ï†ÑÎûµÎ≥¥Îã§ Îçî ÎÜíÏùÄ ÏÑ±Í≥ºÎ•º Ïù¥ÎÅåÏñ¥ ÎÇº Ïàò ÏûàÎäîÏßÄ ÌôïÏù∏.
- Í∞ÄÏÑ§:
    - Í∑ÄÎ¨¥Í∞ÄÏÑ§(H0) : Îëê ÎßàÏºÄÌåÖ Ï†ÑÎûµ Í∞ÑÏùò ÏÑ±Í≥º Ï∞®Ïù¥Îäî ÏóÜÎã§. Ï¶â, ÏÉà Ï†ÑÎûµÍ≥º Í∏∞Ï°¥ Ï†ÑÎûµÏùò ÏÑ±Í≥ºÍ∞Ä ÎèôÏùºÌïòÎã§.
    - ÎåÄÎ¶ΩÍ∞ÄÏÑ§(H1) : ÏÉàÎ°úÏö¥ ÎßàÏºÄÌåÖ Ï†ÑÎûµÏù¥ Í∏∞Ï°¥ Ï†ÑÎûµÎ≥¥Îã§ Îçî ÎÜíÏùÄ ÏÑ±Í≥ºÎ•º Î≥¥Ïù∏Îã§.

#### 2. ÌÖåÏä§Ìä∏ Í∏∞Í∞Ñ ÏÑ§Ï†ï
- weekÏª¨ÎüºÏóêÏÑú 1~4Î°ú ÎêòÏñ¥ÏûàÏùå. Ï¶â, ÌïúÎã¨ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú ÌôïÏù∏ Ìï† Í≤É.

#### 3. ÌÜµÍ≥ÑÏ†Å Î∂ÑÏÑù ÏàòÌñâ
- ÌèâÍ∑† Ï∞®Ïù¥ Î∂ÑÏÑù: SalesInThousandsÏùò ÌèâÍ∑†ÏùÑ ÎπÑÍµêÌïòÏó¨ Í∞Å Í∑∏Î£πÏùò ÏÑ±Í≥º Ï∞®Ïù¥Î•º ÌôïÏù∏Ìï©ÎãàÎã§.
- T-Í≤ÄÏ†ï: Îëê Í∑∏Î£π Í∞ÑÏùò ÏÑ±Í≥º Ï∞®Ïù¥Í∞Ä ÌÜµÍ≥ÑÏ†ÅÏúºÎ°ú Ïú†ÏùòÎØ∏ÌïúÏßÄ ÌôïÏù∏ÌïòÍ∏∞ ÏúÑÌï¥ ÎèÖÎ¶Ω ÌëúÎ≥∏ T-Í≤ÄÏ†ïÏùÑ ÏàòÌñâÌï©ÎãàÎã§.

#### Promotion 1Í≥º Promotion 2 ÏÑ±Í≥º Ï∞®Ïù¥ Î∂ÑÏÑù

````python
import pandas as pd
import scipy.stats as stats

# A/B Í∑∏Î£π ÎÇòÎàÑÍ∏∞
group_a = df[df['Promotion'] == 1]
group_b = df[df['Promotion'] == 2]

# ÌèâÍ∑† Ï∞®Ïù¥ Í≤ÄÏ†ï
t_stat, p_value = stats.ttest_ind(group_a['SalesInThousands'], group_b['SalesInThousands'])

print(f'group_a ÌèâÍ∑†: {group_a['SalesInThousands'].mean()}')
print(f'group_b ÌèâÍ∑†: {group_b['SalesInThousands'].mean()}')
print(f"T-statistic: {t_stat}, P-value: {p_value}")
````

#### Î∂ÑÏÑù
- T-statisticÏù¥ 6.45Î°ú Îß§Ïö∞ ÎÜíÏùÄ Í∞íÏùÑ Í∞ÄÏßÄÍ≥† ÏûàÏäµÎãàÎã§. Ïù¥Îäî Îëê Í∑∏Î£π Í∞ÑÏùò ÌèâÍ∑† Ï∞®Ïù¥Í∞Ä ÏÉÅÎãπÌûà ÌÅ¨Îã§Îäî Í≤ÉÏùÑ ÏùòÎØ∏Ìï©ÎãàÎã§.
- P-valueÍ∞Ä 3.55e-10ÏúºÎ°ú Îß§Ïö∞ ÎÇÆÏùÄ Í∞íÏúºÎ°ú ÌÜµÍ≥ÑÏ†ÅÏúºÎ°ú Ïú†ÏùòÎØ∏ÌïòÎã§Îäî Í≤ÉÏùÑ ÏùòÎØ∏
- Ï¶â, p-value < 0.05Ïù¥ÎØÄÎ°ú Í∑ÄÎ¨¥Í∞ÄÏÑ§ÏùÑ Í∏∞Í∞ÅÌïòÍ≥† ÎåÄÎ¶ΩÍ∞ÄÏÑ§ÏùÑ Ï±ÑÌÉùÌï† Ïàò ÏûàÏäµÎãàÎã§.

#### Í≤∞Î°†
- Í∏∞Ï°¥ Ï†ÑÎûµÏù¥ Îçî ÎÜíÏùÄ ÏÑ±Í≥ºÎ•º Î≥¥ÏòÄÎã§: Group A(Í∏∞Ï°¥ Ï†ÑÎûµ)Ïùò ÌèâÍ∑† Îß§Ï∂úÏù¥ Group B(ÏÉà Ï†ÑÎûµ)Î≥¥Îã§ Ìõ®Ïî¨ ÎÜíÏùÄ Í≤ÉÏúºÎ°ú ÎÇòÌÉÄÎÇ¨ÏäµÎãàÎã§. ÎòêÌïú, ÌÜµÍ≥ÑÏ†ÅÏúºÎ°ú Ïú†ÏùòÎØ∏Ìïú Ï∞®Ïù¥Í∞Ä ÏûàÏùåÏùÑ ÌôïÏù∏ÌñàÏäµÎãàÎã§.

- Ï∂îÏ≤ú: ÏÉà Ï†ÑÎûµ(B Í∑∏Î£π)ÏùÄ Í∏∞Ï°¥ Ï†ÑÎûµ(A Í∑∏Î£π)Î≥¥Îã§ ÎÇÆÏùÄ ÏÑ±Í≥ºÎ•º Î≥¥ÏòÄÍ∏∞ ÎïåÎ¨∏Ïóê, Í∏∞ÏóÖ XYZÎäî Í∏∞Ï°¥ Ï†ÑÎûµÏùÑ Ïú†ÏßÄÌïòÍ±∞ÎÇò ÏÉà Ï†ÑÎûµÏùÑ Í∞úÏÑ†Ìï¥Ïïº Ìï† ÌïÑÏöîÍ∞Ä ÏûàÏùÑ Í≤ÉÏúºÎ°ú Î≥¥ÏûÖÎãàÎã§.

#### Promotion 1Í≥º Promotion 3 ÏÑ±Í≥º Ï∞®Ïù¥ Î∂ÑÏÑù

````python
# A/B Í∑∏Î£π ÎÇòÎàÑÍ∏∞
group_a = df[df['Promotion'] == 1]
group_b = df[df['Promotion'] == 3]

# ÌèâÍ∑† Ï∞®Ïù¥ Í≤ÄÏ†ï
t_stat, p_value = stats.ttest_ind(group_a['SalesInThousands'], group_b['SalesInThousands'])

print(f'group_a ÌèâÍ∑†: {group_a['SalesInThousands'].mean()}')
print(f'group_b ÌèâÍ∑†: {group_b['SalesInThousands'].mean()}')
print(f"T-statistic: {t_stat}, P-value: {p_value}")
````

#### Î∂ÑÏÑù
- T-statistic Í∞íÏùÄ 1.55Î°ú, Îëê Í∑∏Î£π Í∞ÑÏùò ÌèâÍ∑† Ï∞®Ïù¥Í∞Ä Ïñ¥Îäê Ï†ïÎèÑ ÏûàÏßÄÎßå ÌÅ∞ Ï∞®Ïù¥Îäî ÏóÜÎã§Îäî Í≤ÉÏùÑ ÏùòÎØ∏Ìï©ÎãàÎã§.
- P-valueÎäî 0.12Î°ú 0.05Î≥¥Îã§ ÌÅ∞ Í∞íÏúºÎ°ú ÌÜµÍ≥ÑÏ†ÅÏúºÎ°ú Ïú†ÏùòÎØ∏ÌïòÏßÄ ÏïäÏäµÎãàÎã§.
- Ï¶â, Í∑ÄÎ¨¥Í∞ÄÏÑ§ÏùÑ Í∏∞Í∞Å Ìï† Ïàò ÏóÜÍ≥† ÏÉà Ï†ÑÎûµÍ≥º Í∏∞Ï°¥ Ï†ÑÎûµ Í∞ÑÏùò Ï∞®Ïù¥Í∞Ä Ïö∞Ïó∞Ïóê ÏùòÌïú Í≤ÉÏùº Í∞ÄÎä•ÏÑ±Ïù¥ ÌÅΩÎãàÎã§.

#### Í≤∞Î°†
- Îëê Ï†ÑÎûµ Í∞ÑÏùò ÏÑ±Í≥º Ï∞®Ïù¥Îäî ÌÜµÍ≥ÑÏ†ÅÏúºÎ°ú Ïú†ÏùòÎØ∏ÌïòÏßÄ ÏïäÎã§Í≥† Î≥º Ïàò ÏûàÏäµÎãàÎã§.
- Group A (Í∏∞Ï°¥ Ï†ÑÎûµ)ÏôÄ Group B (ÏÉà Ï†ÑÎûµ)Ïùò ÌèâÍ∑† Ï∞®Ïù¥Í∞Ä ÏûàÍ∏¥ ÌïòÏßÄÎßå, Í∑∏ Ï∞®Ïù¥Í∞Ä Ïö∞Ïó∞Ïóê ÏùòÌïú Í≤ÉÏùº Ïàò ÏûàÍ∏∞ ÎïåÎ¨∏Ïóê, ÏÉà Ï†ÑÎûµÏù¥ Í∏∞Ï°¥ Ï†ÑÎûµÎ≥¥Îã§ Îçî Ìö®Í≥ºÏ†ÅÏù¥ÎùºÍ≥† Í≤∞Î°† ÎÇ¥Î¶¨Í∏∞ Ïñ¥Î†µÏäµÎãàÎã§.

#### Promotion 2ÏôÄ Promotion 3 ÏÑ±Í≥º Ï∞®Ïù¥ Î∂ÑÏÑù

````python
# A/B Í∑∏Î£π ÎÇòÎàÑÍ∏∞
group_a = df[df['Promotion'] == 2]
group_b = df[df['Promotion'] == 3]

# ÌèâÍ∑† Ï∞®Ïù¥ Í≤ÄÏ†ï
t_stat, p_value = stats.ttest_ind(group_a['SalesInThousands'], group_b['SalesInThousands'])

print(f'group_a ÌèâÍ∑†: {group_a['SalesInThousands'].mean()}')
print(f'group_b ÌèâÍ∑†: {group_b['SalesInThousands'].mean()}')
print(f"T-statistic: {t_stat}, P-value: {p_value}")
````

#### Î∂ÑÏÑù
- T-statistic Í∞íÏùÄ -4.88Î°ú Îß§Ïö∞ ÎÇÆÏäµÎãàÎã§. Ïù¥Îäî Îëê Í∑∏Î£π Í∞Ñ Ï∞®Ïù¥Í∞Ä ÌÅ∞ Ï∞®Ïù¥Í∞Ä ÏûàÎäî Í≤ÉÏùÑ ÏùòÎØ∏Ìï©ÎãàÎã§.
- P-valueÎäî 1.56e-06Î°ú Îß§Ïö∞ ÏûëÏäµÎãàÎã§. Í∑∏Îü¨ÎØÄÎ°ú Îëê Í∑∏Î£π Í∞ÑÏùò Ï∞®Ïù¥Îäî ÌÜµÍ≥ÑÏ†ÅÏúºÎ°ú Ïú†ÏùòÎØ∏ Ìïú Í≤ÉÏúºÎ°ú Î≥¥ÏûÖÎãàÎã§.
- Ï¶â, Í∑ÄÎ¨¥Í∞ÄÏÑ§ÏùÑ Í∏∞Í∞ÅÌïòÍ≥† ÎåÄÎ¶ΩÍ∞ÄÏÑ§ÏùÑ Ï±ÑÌÉùÌï† Ïàò ÏûàÏäµÎãàÎã§. Í∑∏Îü¨ÎØÄÎ°ú ÏÉà Ï†ÑÎûµÏù¥ Í∏∞Ï°¥ Ï†ÑÎûµÎ≥¥Îã§ Îçî Ï¢ãÏùÄ ÏÑ±Í≥ºÎ•º Î≥¥Ïù∏Îäî Í≤ÉÏûÖÎãàÎã§.
